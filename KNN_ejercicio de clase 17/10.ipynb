```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
ruta_archivo= "Iris.csv"

# Leer el archivo CSV
df = pd.read_csv(ruta_archivo, header=None)
# Asignar nombres a las columnas
df.columns = ['id','sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
# Eliminar la columna id
df = df.drop('id', axis=1)
# Mostrar las primeras filas del DataFrame
print(df.head())
```

Realiza un análisis descriptivo de los datos: histogramas, diagrama de
dispersión. ¿Qué características (features) son relevantes? *texto en cursiva* **texto en negrita**

```
import matplotlib.pyplot as plt
import seaborn as sns
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
# Graficar histogramas de las características agrupadas por tipo de iris
plt.figure(figsize=(10, 5))
for i, feature in enumerate(features):
plt.subplot(2, 2, i + 1)
sns.histplot(data=df, x=feature, hue='species', bins=15)
plt.title(f'Histograma de {feature}')
plt.xlabel(feature)
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```
# Asegurarse de que la columna 'species' es categórica
df['species'] = df['species'].astype('category')
# Crear la matriz de gráficos de dispersión
sns.pairplot(df, hue='species', palette='viridis')
plt.suptitle('Matriz de Dispersión del Dataset Iris',y=1)
plt.show()
```

```
from sklearn.model_selection import train_test_split
# Dividir el conjunto de datos en entrenamiento 50% y test 50%
X = df.drop('species', axis=1)
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5 ,   #aquí se pone el % de división de la data
random_state=42 , stratify=y)    #randon con semilla 42, el stratify =y es para que los datos se alineen al vector tarjet
```

3) Implementa un KNN usando todas las características, con K=3

```
from sklearn.neighbors import KNeighborsClassifier
#Entrenamos el modelo con 3 vecinos usando los datos de entrenamiento
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train) # punto fit ejecuta el entrenamiento, se le dice cual es el vector y de entrenamiento el primer argumento es X_train y el Segundo es Y_train
```

guarda el dataset como memoria para la predicción


Métricas del modelo
Accuracy=correct predictions/all predictions
Es en cuantos valores estamos acetardando

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
#Evaluamos el rendimiento del modelo usando los datos del test
y_test_pred = knn.predict(X_test)    #punto predict, es el vector de las obs del test
accuracy = accuracy_score(y_test,y_test_pred)
print(f'Accuracy del modelo KNN: {accuracy:.2f}')
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Calculamos la matriz de confusión (objeto)
cm = confusion_matrix(y_test, y_test_pred, labels=df['species'].unique())
# Graficamos la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=df['species'].unique())
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
disp.plot(cmap='viridis', values_format='d')
plt.title('Confusion Matrix for K-Nearest Neighbors Classifier')
plt.show()
```
